# Ejercicio 1: bucle while

```python
new_planet = ''
planets = []
while new_planet.lower() != 'fin':
    if new_planet:
        planets.append(new_planet)
    new_planet = input('Escribe un planeta ')
```

# Ejercicio 2: Ciclo for

```python
print('Planetas ingresados:')
for planet in planets:
    print(planet)
```
